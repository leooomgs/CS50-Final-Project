```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import re
import string
```

```python
data_fake = pd.read_csv('Fake (1).csv')
data_true = pd.read_csv('True.csv')
```

```python
data_fake["class"] = 0 
data_true['class'] = 1
```

```python
data_fake_manual_testing = data_fake.tail(10)
for i in range(23480, 23470,-1):
    data_fake.drop([i], axis = 0, inplace = True)

data_true_manual_testing = data_true.tail(10)
for i in range(21416, 21406,-1):
    data_true.drop([i], axis = 0, inplace = True)
```

```python
data_merge = pd.concat([data_fake, data_true], axis = 0)
data_merge.head(10)
```

```python
data_merge.columns
```

```python
data = data_merge.drop(['title','subject', 'date'], axis = 1)
```

```python
data.isnull().sum()
```

```python
data = data.sample(frac = 1)
```

```python
data.head()
```

```python
data.reset_index(inplace = True)
data.drop(['index'], axis = 1, inplace = True)
```

```python
data.columns
```

```python
data.head()
```

```python
def wordopt(text):
    text = text.lower()
    text = re.sub(r'\(reuters\)\s*-\s*', '', text)
    text = re.sub(r'\[.*?\]', ' ', text)
    text = re.sub(r'\\W', ' ', text)
    text = re.sub(r'https?://\S+|www\.\S+', ' ', text)
    text = re.sub(r'<.*?>+', ' ', text)
    text = re.sub(f'[{re.escape(string.punctuation)}]', ' ', text)
    text = re.sub(r'\n', ' ', text)
    text = re.sub(r'\w*\d\w*', ' ', text)
    return text
```

```python
data['text'] = data['text'].apply(wordopt)
```

```python
x = data['text']
y = data['class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer


vectorization = TfidfVectorizer()


xv_train = vectorization.fit_transform(x_train)


xv_test = vectorization.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression()
LR.fit(xv_train,y_train)
```

```python
prediction_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
print(classification_report(y_test, prediction_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier

DT = DecisionTreeClassifier()
DT.fit(xv_train, y_train)
```

```python
prediction_dt = DT.predict(xv_test)
```

```python
DT.score(xv_test, y_test)
```

```python
print(classification_report(y_test, prediction_dt))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

GB = GradientBoostingClassifier(random_state = 0)
GB.fit(xv_train, y_train)
```

```python
predict_gb = GB.predict(xv_test)
```

```python
GB.score(xv_test, y_test)
```

```python
print(classification_report(y_test, predict_gb))
```

```python
def output_label(n):
    if n == 0:
        return "Fake News"
    elif n == 1:
        return "Not A Fake News"

# Funktion für die manuelle Eingabe und Testung eines neuen Textes
def manual_testing(news):
    testing_news = {"text": [news]}  # Verpacken des Textes in ein Dictionary
    new_def_test = pd.DataFrame(testing_news)  # Erstellung eines DataFrame
    new_def_test["text"] = new_def_test["text"].apply(wordopt)  # Anwendung der Preprocessing-Funktion
    new_x_test = new_def_test["text"]  # Auswahl der Textspalte
    new_xv_test = vectorization.transform(new_x_test)  # Umwandlung des Textes in Vektoren
    
    # Vorhersagen mit den verschiedenen Modellen
    pred_LR = LR.predict(new_xv_test)
    pred_DT = DT.predict(new_xv_test)
    pred_GB = GB.predict(new_xv_test)
    
    # Rückgabe der Vorhersagen als Labels
    return (output_label(pred_LR[0]), 
            output_label(pred_DT[0]), 
            output_label(pred_GB[0]))
```
